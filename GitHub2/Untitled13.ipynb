```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import umap
from rdkit import Chem
from rdkit.Chem import AllChem, Descriptors, rdMolDescriptors
import os
from typing import Dict, List, Optional
import pandas as pd
from collections import defaultdict

def load_embeddings(filepath):
    """Load embeddings and molecule data from file"""
    with open(filepath, 'rb') as f:
        data = pickle.load(f)
    
    embeddings = data['embeddings']
    molecule_data = data.get('molecule_data', None)
    
    return embeddings, molecule_data

def rebuild_molecules_from_graph_data(molecule_data_list):
    """Reconstruct RDKit molecules from graph data (simplified version)"""
    # Maps for atom and bond features
    atom_list = list(range(1, 119))
    chirality_list = [
        Chem.rdchem.ChiralType.CHI_UNSPECIFIED,
        Chem.rdchem.ChiralType.CHI_TETRAHEDRAL_CW,
        Chem.rdchem.ChiralType.CHI_TETRAHEDRAL_CCW,
        Chem.rdchem.ChiralType.CHI_OTHER
    ]
    bond_list = [
        Chem.rdchem.BondType.SINGLE,
        Chem.rdchem.BondType.DOUBLE, 
        Chem.rdchem.BondType.TRIPLE,
        Chem.rdchem.BondType.AROMATIC
    ]
    
    molecules = []
    for i, mol_data in enumerate(molecule_data_list):
        try:
            # Create empty editable molecule
            mol = Chem.RWMol()
            
            # Add atoms based on x_cat data
            x_cat = mol_data.get('x_cat')
            if x_cat is None:
                molecules.append(None)
                continue
                
            for atom_features in x_cat:
                atomic_num = atom_list[atom_features[0]]
                atom = Chem.Atom(atomic_num)
                atom.SetChiralTag(chirality_list[atom_features[1]])
                mol.AddAtom(atom)
            
            # Add bonds based on edge_index and edge_attr
            edge_index = mol_data.get('edge_index')
            edge_attr = mol_data.get('edge_attr')
            
            if edge_index is None or edge_attr is None:
                molecules.append(Chem.Mol(mol))
                continue
                
            # Process only in one direction (undirected edges appear twice)
            processed_edges = set()
            
            for i in range(len(edge_index[0])):
                start, end = edge_index[0][i], edge_index[1][i]
                if (start, end) in processed_edges or (end, start) in processed_edges:
                    continue
                    
                # Add the bond
                bond_type = bond_list[int(edge_attr[i][0])]
                mol.AddBond(start, end, bond_type)
                processed_edges.add((start, end))
            
            # Convert to molecule and add hydrogen
            final_mol = Chem.Mol(mol)
            Chem.SanitizeMol(final_mol)
            final_mol = Chem.AddHs(final_mol)
            AllChem.Compute2DCoords(final_mol)
            molecules.append(final_mol)
            
        except Exception as e:
            if i % 1000 == 0:
                print(f"Error rebuilding molecule {i}: {e}")
            molecules.append(None)
    
    return molecules

def identify_functional_groups(molecules):
    """Identify functional groups in molecules using RDKit's functional group filtering"""
    from rdkit.Chem import MolFromSmarts
    
    # Define common functional groups with SMARTS patterns
    functional_groups = {
        'carboxylic_acid': '[CX3](=O)[OX2H1]',
        'ester': '[CX3](=O)[OX2][CX4]',
        'amide': '[NX3][CX3](=[OX1])',
        'amine': '[NX3;H2,H1,H0;!$(NC=O)]',
        'hydroxyl': '[OX2H]',
        'carbonyl': '[CX3]=[OX1]',
        'alkene': '[CX3]=[CX3]',
        'alkyne': '[CX2]#[CX2]',
        'aromatic': 'c1ccccc1',
        'heterocycle': '[a;!c]',
        'phenol': '[OX2H][cX3]:[cX3]',
        'nitro': '[NX3](=O)=O',
        'nitrile': '[NX1]#[CX2]',
        'halogen': '[F,Cl,Br,I]',
        'sulfide': '[#16X2][#6]',
        'sulfonyl': '[#16X4](=[OX1])(=[OX1])',
        'phosphate': '[P,p][OX2][CX4]'
    }
    
    # Compile patterns
    patterns = {name: MolFromSmarts(smarts) for name, smarts in functional_groups.items()}
    
    # Initialize results
    group_counts = defaultdict(list)
    
    for mol in molecules:
        if mol is None:
            # Add zeros for all groups if molecule is None
            for group in functional_groups:
                group_counts[group].append(0)
            continue
        
        # Count occurrences of each functional group
        for group_name, pattern in patterns.items():
            if pattern is not None:
                matches = mol.GetSubstructMatches(pattern)
                group_counts[group_name].append(len(matches))
            else:
                group_counts[group_name].append(0)
    
    # Convert to numpy arrays
    for key in group_counts:
        group_counts[key] = np.array(group_counts[key])
    
    return group_counts

def reduce_dimensions(embeddings, method='umap', n_components=2, **kwargs):
    """Reduce dimensionality of embeddings for visualization"""
    if method.lower() == 'tsne':
        reducer = TSNE(n_components=n_components, random_state=42, **kwargs)
    elif method.lower() == 'umap':
        reducer = umap.UMAP(n_components=n_components, random_state=42, **kwargs)
    elif method.lower() == 'pca':
        reducer = PCA(n_components=n_components, **kwargs)
    else:
        raise ValueError(f"Unknown dimensionality reduction method: {method}")
    
    reduced_embeddings = reducer.fit_transform(embeddings)
    return reduced_embeddings

def cluster_embeddings(embeddings_2d, n_clusters=8):
    """Cluster the embeddings using KMeans"""
    # Apply KMeans clustering
    kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
    clusters = kmeans.fit_predict(embeddings_2d)
    
    return clusters

def analyze_functional_groups_by_cluster(functional_groups, clusters):
    """Analyze distribution of functional groups across clusters"""
    # Get unique clusters
    unique_clusters = np.unique(clusters)
    n_clusters = len(unique_clusters)
    
    # Create a dictionary to store average counts per cluster
    avg_counts = {}
    
    # Calculate average count for each functional group in each cluster
    for group_name, counts in functional_groups.items():
        avg_counts[group_name] = []
        
        for cluster_id in unique_clusters:
            # Get indices for this cluster
            cluster_mask = clusters == cluster_id
            
            # Calculate average count
            if np.sum(cluster_mask) > 0:
                avg = np.mean(counts[cluster_mask])
            else:
                avg = 0
                
            avg_counts[group_name].append(avg)
    
    # Convert to DataFrame for easier visualization
    df = pd.DataFrame(avg_counts)
    df.index = [f'Cluster {i}' for i in unique_clusters]
    
    return df

def analyze_property_distributions(molecules, clusters):
    """Analyze distributions of molecular properties across clusters"""
    from rdkit.Chem import Descriptors, Lipinski
    
    # Define properties to analyze
    property_calculators = {
        'Molecular Weight': Descriptors.MolWt,
        'LogP': Descriptors.MolLogP,
        'TPSA': Descriptors.TPSA,
        'Number of Rotatable Bonds': Descriptors.NumRotatableBonds,
        'Number of H-Bond Donors': Descriptors.NumHDonors,
        'Number of H-Bond Acceptors': Descriptors.NumHAcceptors,
        'Number of Rings': Descriptors.RingCount,
        'Number of Aromatic Rings': lambda m: rdMolDescriptors.CalcNumAromaticRings(m),
        'Number of Heavy Atoms': Descriptors.HeavyAtomCount
    }
    
    # Initialize results dictionary
    property_values = {name: [] for name in property_calculators}
    
    # Calculate properties for each molecule
    for mol in molecules:
        if mol is None:
            for name in property_calculators:
                property_values[name].append(None)
            continue
            
        for name, calculator in property_calculators.items():
            try:
                value = calculator(mol)
                property_values[name].append(value)
            except:
                property_values[name].append(None)
    
    # Convert to numpy arrays with proper handling of None values
    for name in property_values:
        values = np.array(property_values[name], dtype=float)
        property_values[name] = values
    
    # Get unique clusters
    unique_clusters = np.unique(clusters)
    
    # Calculate statistics per cluster
    cluster_stats = []
    
    for cluster_id in unique_clusters:
        cluster_mask = clusters == cluster_id
        
        # Skip if no molecules in cluster
        if np.sum(cluster_mask) == 0:
            continue
            
        # Calculate statistics for this cluster
        stats = {'Cluster': f'Cluster {cluster_id}', 'Count': np.sum(cluster_mask)}
        
        for name, values in property_values.items():
            cluster_values = values[cluster_mask]
            # Filter out None/NaN values
            cluster_values = cluster_values[~np.isnan(cluster_values)]
            
            if len(cluster_values) > 0:
                stats[f'{name} (Mean)'] = np.mean(cluster_values)
                stats[f'{name} (Std)'] = np.std(cluster_values)
                stats[f'{name} (Min)'] = np.min(cluster_values)
                stats[f'{name} (Max)'] = np.max(cluster_values)
            else:
                stats[f'{name} (Mean)'] = np.nan
                stats[f'{name} (Std)'] = np.nan
                stats[f'{name} (Min)'] = np.nan
                stats[f'{name} (Max)'] = np.nan
                
        cluster_stats.append(stats)
    
    # Convert to DataFrame
    df = pd.DataFrame(cluster_stats)
    
    return df, property_values

def plot_functional_group_heatmap(func_group_df, save_path=None):
    """Plot heatmap of functional group distribution across clusters"""
    plt.figure(figsize=(14, 10))
    
    # Create heatmap
    ax = sns.heatmap(func_group_df, annot=True, cmap='viridis', fmt='.2f', linewidths=0.5)
    
    # Set labels
    plt.title('Average Functional Group Distribution Across Clusters', fontsize=16)
    plt.xlabel('Functional Groups', fontsize=14)
    plt.ylabel('Clusters', fontsize=14)
    
    # Rotate x-axis labels for better readability
    plt.xticks(rotation=45, ha='right')
    
    plt.tight_layout()
    
    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
        print(f"Saved functional group heatmap to {save_path}")
    
    plt.show()

def plot_property_distributions(property_values, clusters, save_dir=None):
    """Plot distributions of molecular properties across clusters"""
    # Create save directory if needed
    if save_dir:
        os.makedirs(save_dir, exist_ok=True)
    
    # Get unique clusters
    unique_clusters = np.unique(clusters)
    n_clusters = len(unique_clusters)
    
    # Set up colors for clusters
    colors = plt.cm.viridis(np.linspace(0, 1, n_clusters))
    
    # Plot each property
    for property_name, values in property_values.items():
        plt.figure(figsize=(12, 8))
        
        for i, cluster_id in enumerate(unique_clusters):
            cluster_mask = clusters == cluster_id
            cluster_values = values[cluster_mask]
            
            # Filter out None/NaN values
            cluster_values = cluster_values[~np.isnan(cluster_values)]
            
            if len(cluster_values) > 0:
                # Plot KDE (density plot)
                sns.kdeplot(cluster_values, color=colors[i], label=f'Cluster {cluster_id}')
        
        plt.title(f'Distribution of {property_name} Across Clusters', fontsize=16)
        plt.xlabel(property_name, fontsize=14)
        plt.ylabel('Density', fontsize=14)
        plt.legend()
        
        plt.tight_layout()
        
        if save_dir:
            # Create a clean filename
            clean_name = property_name.lower().replace(' ', '_')
            plt.savefig(os.path.join(save_dir, f'{clean_name}_distribution.png'), dpi=300)
        
        plt.show()

def export_cluster_analysis(cluster_stats_df, save_path=None):
    """Export cluster analysis to CSV file"""
    if save_path:
        cluster_stats_df.to_csv(save_path, index=False)
        print(f"Exported cluster statistics to {save_path}")
    
    return cluster_stats_df

def main():
    # Create output directory
    output_dir = './analysis_results'
    os.makedirs(output_dir, exist_ok=True)
    
    # Load the embeddings (replace with your actual path)
    embeddings_path = './embeddings/final_embeddings_molecules_20250308_221455.pkl'
    print(f"Loading embeddings from: {embeddings_path}")
    embeddings, molecule_data = load_embeddings(embeddings_path)
    
    print(f"Loaded {len(embeddings)} embeddings")
    
    # Reduce dimensions using UMAP
    print("Reducing dimensionality with UMAP...")
    embeddings_2d = reduce_dimensions(
        embeddings, 
        method='umap', 
        n_neighbors=15, 
        min_dist=0.1, 
        metric='cosine'
    )
    
    # Cluster the embeddings
    print("Clustering embeddings...")
    n_clusters = 10  # Adjust based on your data
    clusters = cluster_embeddings(embeddings_2d, n_clusters=n_clusters)
    
    # Rebuild molecules
    print("Rebuilding molecules from graph data...")
    molecules = rebuild_molecules_from_graph_data(molecule_data)
    print(f"Successfully rebuilt {sum(1 for m in molecules if m is not None)} molecules")
    
    # Identify functional groups
    print("Identifying functional groups...")
    func_groups = identify_functional_groups(molecules)
    
    # Analyze functional groups by cluster
    print("Analyzing functional group distribution across clusters...")
    func_group_df = analyze_functional_groups_by_cluster(func_groups, clusters)
    
    # Plot functional group heatmap
    print("Plotting functional group heatmap...")
    plot_functional_group_heatmap(
        func_group_df, 
        save_path=os.path.join(output_dir, 'functional_group_heatmap.png')
    )
    
    # Analyze property distributions
    print("Analyzing molecular property distributions...")
    cluster_stats_df, property_values = analyze_property_distributions(molecules, clusters)
    
    # Export cluster statistics
    print("Exporting cluster statistics...")
    export_cluster_analysis(
        cluster_stats_df, 
        save_path=os.path.join(output_dir, 'cluster_statistics.csv')
    )
    
    # Plot property distributions
    print("Plotting property distributions...")
    plot_property_distributions(
        property_values, 
        clusters, 
        save_dir=os.path.join(output_dir, 'property_distributions')
    )
    
    print("Analysis complete!")

if __name__ == "__main__":
    main()
```
